```python
Oct_2_Group
```

```python
def calculate_points(column):
    column['Deposit - Withdrawal'] = column['Deposit Amount'] - column['Withdrawal Amount']
    column['Max Difference'] = column['Deposit - Withdrawal'].clip(lower=0)
    column['Points_Deposit_Withdrawal'] = 0.001 * column['Max Difference']

    column['Points_Games_Played'] = 0.2 * column['Games Played']

    column['Loyalty Points'] = column['Points_Deposit_Withdrawal'] + column['Points_Games_Played']

    column.drop(['Deposit - Withdrawal', 'Max Difference', 'Points_Deposit_Withdrawal', 'Points_Games_Played'], axis=1, inplace=True)

    return column
```

```python
Oct_2_Points = calculate_points(Oct_2_Group)
```

### 2. Calculate overall loyalty points earned and rank players on the basis of loyalty points in the month of October.  In case of tie, number of games played should be taken as the next criteria for ranking.

```python
df.head(5)
```

```python
wf.head(5)
```

```python
gf.head(5)
```

```python
overall = pd.concat([df,wf,gf],ignore_index=True)
```

```python
overall
```

```python
Overall_Groupby = overall.groupby('User Id')[['Deposit Amount', 'Withdrawal Amount', 'Games Played']].sum().reset_index()
```

```python
Overall_Groupby
```

```python
Overall_Points = calculate_points(Overall_Groupby)
```

```python
Overall_Points
```

```python
Overall_Points
```

```python
ranks = []

for index, row in Overall_Points.iterrows():
    rank = 1
    for _, other_row in Overall_Points.iterrows():
        if index == _:
            continue
        if row['Loyalty Points'] < other_row['Loyalty Points']:
            rank += 1
        elif row['Loyalty Points'] == other_row['Loyalty Points']:
            if row['Games Played'] < other_row['Games Played']:
                rank += 1
    ranks.append(rank)

Overall_Points['Rank'] = ranks

Rank_wise_Players = Overall_Points.sort_values(by='Rank')
rank_col = Rank_wise_Players.pop('Rank')
Rank_wise_Players.insert(0,'Rank',rank_col)
```

## Part C

Would you say the loyalty point formula is fair or unfair?

Can you suggest any way to make the loyalty point formula more robust?


The current loyalty program can be considered reasonably fair because:

It awards points for multiple positive behaviors like deposits, withdrawals, games played rather than just one metric.
There is transparency in how points are calculated based on defined formulas for each action. Players can estimate their points.
Rewards active and engaged players who use the platform more. More usage = more points.

We Can make loyalty points more robust by creating columns as Customer value.
where points are distributed based on the value contributed by customer

Increase weightage for withdrawals and games played compared to deposits.

Introduce levels and badges.

Use Pubic leaderboards and tier progress to motivate players so that they can engage more.
